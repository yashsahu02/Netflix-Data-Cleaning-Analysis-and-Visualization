### date_added

```python
df['date_added'].dtype
```

```python
df['date_added'].sample(5)
```

Here the type is object so let's convert it to pandas date time

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['date_added'].dtype
```

```python
df.head(2)
```

```python
### Extracting year, month, day from date_added
df['date_added_year'] = df['date_added'].dt.year
df['date_added_month'] = df['date_added'].dt.month
df['date_added_month_name'] = df['date_added'].dt.month_name()
df['date_added_day'] = df['date_added'].dt.day
```

```python
df.sample(5)
```

```python
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
df['date_added_year'].value_counts().head(5).plot(kind='bar')
plt.title("Bar Chart 'date_added_year' feature (Top 5)")

plt.subplot(1,2,2)
df['date_added_year'].value_counts().head(5).plot(kind='pie', autopct = "%.2f%%")
plt.title("Pie Chart 'date_added_year' feature (Top 5)")
plt.show()
```

```python
df['date_added_year'].value_counts().tail(5) ## least occuring years in date_added
```

#### Observation
- These are top 5 most occuring years and maximum content added in year 2019 (approx 2000).
- Minimum content added in 2010 occuring only 1 time.

```python
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
df['date_added_month_name'].value_counts().plot(kind='bar')
plt.title("Bar Chart 'date_added_month_name' feature")

plt.subplot(1,2,2)
df['date_added_month_name'].value_counts().plot(kind='pie', autopct = "%.2f%%")
plt.title("Pie Chart 'date_added_month_name' feature")
plt.show()
```

#### Observation
- July is highly active as date_added month (more than 800) and february is lowest active (approx 560)
